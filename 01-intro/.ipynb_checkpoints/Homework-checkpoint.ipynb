```python
df = pd.read_parquet('https://d37ci6vzurychx.cloudfront.net/trip-data/yellow_tripdata_2025-01.parquet')
```

```python
#frames = [df_jan,df_feb,df_mar,df_apr,df_mar]
```

```python
df['tpep_dropoff_dt'] = pd.to_datetime(df.tpep_dropoff_datetime)
df['tpep_pickup_dt'] = pd.to_datetime(df.tpep_pickup_datetime)
```

```python
df['duration'] = df.tpep_dropoff_dt - df.tpep_pickup_dt
```

```python
df
```

```python
# calculate trip duration in minutes: 
df.duration = df.duration.apply(lambda td: td.total_seconds()/60)
```

```python
df
```

```python
df.duration.describe()
```

```python
df_positive = df[(df['duration'] > 0) & (df['duration'] < 1440)]
```

```python
df_positive.duration.describe()
```

```python
df_positive = df_positive.copy()
df_positive['fare_per_mile'] = df_positive['fare_amount'].astype(float) / df_positive['trip_distance'].astype(float)
```

```python
df_positive.fare_per_mile
```

```python
# Show how fares are distributed 
# if it's too little, the distance might be an outlier
sns.displot(df_positive.fare_per_mile)

plt.show()
```
